```python
from datetime import datetime, date, timedelta
import schedule
import time

loop = 1
def SetSalePriceForProductsAtTheWeekend():
    print("Set Sale Schedule is Auto Working")
    start_sale_date = str(date.today())+"T00:00:00"
    end_sale_date = str(date.today())+"T23:59:59"
    try:
        if (datetime.today().weekday() == 5):
            page = 1 
            while True:
                products = wcapi.get('products', params={'per_page': 100, 'page': page}).json()
                page += 1
                for product in products:
                    #print(product['id'])
                    data = {
                        "sale_price": str(float(product['regular_price'])*0.9),
                        "date_on_sale_from": start_sale_date,
                        "date_on_sale_to": end_sale_date
                    }
                    link = "products/" + str(product['id'])
                    wcapi.put(link, data).json()
                if not products:
                    break
    except:
        pass

schedule.every(1).days.do(SetSalePriceForProductsAtTheWeekend)

while (loop <= 5):
    schedule.run_pending()
    time.sleep(1)
    loop = loop + 1
```
